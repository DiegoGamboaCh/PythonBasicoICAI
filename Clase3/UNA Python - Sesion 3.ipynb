# Control de Flujos

- El flujo de control se refiere al orden en que se debe ejecutar el programa
- Las declaraciones de flujo de control rompen el orden general de ejecución de arriba a abajo al incluir la toma de decisiones, el bucle y más
- Esto permite que el programa ejecute primero un bloque de código específico en función de las condiciones utilizadas

Ver referencias sobre la control de flujo en sitio oficial
- https://docs.python.org/3/tutorial/controlflow.html#
- https://docs.python.org/3/reference/compound_stmts.html#



## Estatuto IF

- Usada para una ejecución condicional
- Con esta sentencia se busca seleccionar la primera  condición verdadera en  **if**  o **elif** y con ello ejecutar el bloque de código respectivo
- En los casos donde ninguna de las condiciones es verdadera se ejecuta el código dentro del **else**

#### Uso de estatuto if

```python
# if básico
a = 5
if (a < 10):
    print ('5 es menor que  10')
print ('Estatuto fuera del if ')
```

```python
# if-else
i = 20
if (i < 15):
    print ('i  es menor que  15')

else:
    print ('i es mayor que  15')
print ('Estatuto fuera del if ')
```

```python
#if-elif

a = 50

if (a == 20):
    print ('el valor es 20')

elif (a == 30):
    print ('el valor es 30')

elif (a == 40):
    print ('el valor es 40')

else:
    print ('el valor es mayor a  40')
```

#### Anidamiento

```python
a = 20
if (a >= 20):
    if (a <=  25):
        print ('a esta entre  20 y  25')
    else:
        print ('a is mayor  25')
else:
    print ('a es menor a   20')
```

## Estatuto WHILE


- El ciclo **while** en Python permite ejecutar un bloque de código repetidamente mientras se cumpla una condición, la cual se evalúa en cada iteración del ciclo
- Se considera al while como un bucle con número de iteraciones no definidas

### while básico

```python
# while sencillo

x = 5
while x > 0:
    x -=1
    print(x)
```

### while-else 

```python
# while-else
x = 5
while x > 0:
    x -=1
    print(x)
else:
    print("El ciclo ha finalizado")
```

### while-break

```python
# while-break
x = 5
while True:
    x -= 1
    print(x)
    if x == 0:
        break
else:
    # El print no se ejecuta
    print("Fin del ciclo")
```

### while-continue

```python
# interrupción
# finaliza la iteración en curso y transfiere el control a la parte superior del ciclo
# y la condición del ciclo se evalúa nuevamente. Si la condición es verdadera,
# entonces tiene lugar la siguiente iteración
a = 0
while a <5:
    a += 1
    if a == 3:
        continue
    print(a)
```

### while anidado

```python
# while anidados
i = 0
j = 0
while i < 3:
    while j < 3:
        print(i,j)
        j += 1
    i += 1
    j = 0
```

```python
# while anidado
i, j, k = 0, 0, 0  # observe como realizamos esta asiganción
while i < 3:
    while j < 3:
        while k < 3:
            print(i,j,k)
            k += 1
            j += 1
        k = 0
    i += 1
    j = 0
```

### clásicos en programación

```python
# árbol de navidad
z = 7
x = 1
while z > 0:
    print(' ' * z + '*' * x + ' ' * z)
    x+=2
    z-=1
```

```python
# Fibonacci
a, b = 0, 1 # observe esta forma de asignación
while b < 25:
    print(b)
    a, b = b, a + b
```

## Estatuto FOR

- El for es un tipo de bucle, parecido al while pero con ciertas diferencias
- Las diferencias principales son s que el número de iteraciones de un for esta definido y  la condición de parada la define un iterable

### for básico

```python
# for sencillo
for i in range(0, 5):
    print(i)
```

```python
# porqué pudimos hacer esto? ver https://docs.python.org/3/tutorial/classes.html#iterators
for i in "Hola":
    print(i)
```

```python
# usando una lista sencilla
lista = [5, 6, 3, 2]
for i in lista:
    print(i)
```

```python
# usando una matriz
mlista = [[56, 34, 1],
         [12, 4, 5],
         [9, 4, 3]]
for i in mlista:
    print(i)
```

### for anidado

```python
# for anidado
for i in mlista:
    for j in i:
        print(j)
```

### caso de for

```python
# iteración saltandonos elementos
texto = "Bienvenidos"
for i in texto[::2]:
    print(i)
```

### for y comprehensions lists

```python
# for usado con comprehensions lists
print(sum(i for i in range(10)))
```
