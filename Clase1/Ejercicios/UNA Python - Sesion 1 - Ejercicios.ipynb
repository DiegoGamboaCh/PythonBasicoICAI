## Problemas

Un tienda vende galletas artesanales  a 99.99 colones  cada una.Sin embargo la galleta  que no es el día tiene un descuento del 50%. Escriba el código  que comience leyendo el número de galletas vendidas que no son del día. Después el código  debe mostrar el precio habitual de una galleta, el descuento que se le hace por no ser fresca y el costo  final a pagar por las galletas que no fueron del día.

Una tienda  tiene mucho éxito en dos de sus productos: teclados y mouse. Suele hacer venta por correo y la empresa de logística les cobra por peso de cada paquete así que deben calcular el peso de los teclados y mouse que saldrán en cada paquete a demanda. Cada teclado pesa 200 g y cada mouse 75 g. Escribir el código  que lea el número de teclados y mouse vendidos en el último pedido y calcule el peso total del paquete que será enviado.
